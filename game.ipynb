```python
#CODE COURTESY OF Kylie Ying: https://www.youtube.com/watch?v=8ext9G7xspg
import sys
sys.setrecursionlimit(3000)

from player import HumanPlayer, RandomComputerPlayer, SmartComputerPlayer

class TicTacToe:
    def __init__(self):
        self.board = [' ' for _ in range(9)]    #Creates a 3x3 list board
        self.current_winner = None #Keeps track of whom won
        
    def print_board(self):
        #cycles through the board: first row as 012, second as 345, third as 678
        for row in [self.board[i*3:(i+1)*3] for i in range(3)]:
            print('| ' + ' | '.join(row) + ' |')
    
    @staticmethod
    def print_board_nums():
        number_board = [[str(i) for i in range(j*3, (j+1)*3)] for j in range(3)]
        for row in number_board:
            print('| ' + ' | '.join(row) + ' |')
    
    def available_moves(self):
       # moves = [return i for i, spot in enumerate(self.board) if spot == ' ']   #Cleaner code than below
        moves = []
        for (i, spot) in enumerate(self.board):  # ['x', 'x', 'o']  is returned as [(0, 'x'), (1, 'x'), (2, 'o')] 
            if spot == ' ':
                moves.append(i)  #Adds the number of the spot if it's empty
        return moves
    
    def empty_squares(self):
        #Returns True if there are still empty squares
        return ' ' in self.board
    
    def num_empty_squares(self):
        #Returns the number of empty squares
        #return len(self.available_moves())
        return self.board.count(' ')
                        
    def winner(self, square, letter):
        #Function to check the winner of the game
        #Checking if there are 3 in a row, column, or diagonally
        #Checking rows
       # row_index = square // 3
       # row = self.board[row_index*3 : (row_index + 1)*3]    #Fetches the rows as a list from the board
        row1 = [self.board[i] for i in [0, 1, 2]]
        if all(spot == letter for spot in row1):    #Checking whether the spot equalls the letter in the row1
            return True
        row2 = [self.board[i] for i in [3, 4, 5]]
        if all(spot == letter for spot in row2):    #Checking whether the spot equalls the letter in the row2
            return True
        row3 = [self.board[i] for i in [6, 7, 8]]
        if all(spot == letter for spot in row3):    #Checking whether the spot equalls the letter in the row3
            return True
                        
        #Checking columns
      #  col_index = square % 3
      #  collumn = [self.board[col_index + i*3] for i in range(3)]    #Fetches everything in a column that we move to
        column1 = [self.board[i] for i in [0, 3, 6]]
        if all(spot == letter for spot in column1):    #Checking whether the spot equalls the letter in the column1
            return True
        column2 = [self.board[i] for i in [1, 4, 7]]
        if all(spot == letter for spot in column2):    #Checking whether the spot equalls the letter in the column2
            return True
        column3 = [self.board[i] for i in [2, 5, 8]]
        if all(spot == letter for spot in column3):    #Checking whether the spot equalls the letter in the column3
            return True
        
        #Checking diagonals
        #Can only be two choices: 0 4 8, or 2 4 6
        if square % 2 == 0:
            diagonal1 = [self.board[i] for i in [0, 4, 8]]
            if all(spot == letter for spot in diagonal1):    #Checking whether the spot equalls the letter in the first diaginal option
                return True
            diagonal2 = [self.board[i] for i in [2, 4, 6]]
            if all(spot == letter for spot in diagonal2):    #Checking whether the spot equalls the letter in the second diagonal option
                return True
                        
        #If there is no winner, return False
        return False
    
    def make_move(self, square, letter):
        #Function to make a move for any player
        #If the move is valid, assign as True and assign the square to the letter 'X' or 'O'
        #Else assign as False
        if self.board[square] == ' ':
            self.board[square] == letter
            if self.winner(square,letter):
                current_winner = letter
            return True
        else:
            return False
                        
def play(game, x_player, o_player, print_game=True):
    #Returns the winner of the game as a letter, or None for a tie
    if print_game:
        game.print_board_nums()
    
    letter = 'X'   #Assigning a starting letter
    #Using a while loop until there are no more empty squares
    #Once there is a winner, the loop will be broken                     
    while game.empty_squares():
        #Gets the move of an appropriate player                
        if letter == 'O':
            square = o_player.get_move(game)
        else:
            square = x_player.get_move(game)
                        
        #Defining a function to make a move
        if game.make_move(square, letter):
            print(letter + f' makes a move to square {square}')
            game.board[square] = letter
            game.print_board()
            if game.winner(square, letter):
                print(letter + " is the winner. Game Over")
                break
            print('')    #Empty line
                        
         #   if game.current_winner():
         #       if print_game():
         #           print(letter + "is the winner.")
         #       return letter
        
        #Alternating letters
       # letter = '0' if letter == 'X' else letter = 'X'    #If the letter was X, change it to O, else if it's O, change it to X
        if letter == 'O':
            letter = 'X'
        else:
            letter = 'O'
        #The Win check in TicTacToe happens after a player makes a move, therefore it will be appended into the function make_move()
    else:
        print("Both Players tied. Game Over")

x_player = HumanPlayer('X')
o_player = SmartComputerPlayer('O')
t = TicTacToe()
play(t, x_player, o_player, print_game=True)
```
